```python
from urlcheck_smith import SiteClassifier, UrlRecord


def classify_single_url(url: str, *, rules_path: str | None = None, explain: bool = False) -> dict:
    classifier = SiteClassifier(
        rules_path=rules_path,
        explain=explain,
        normalize_domain=True,
    )

    rec = UrlRecord(url=url)
    rec = classifier.classify([rec])[0]

    result = {
        "url": rec.url,
        "base_url": rec.base_url,
        "category": rec.category,
        "trust_tier": rec.trust_tier,
    }

    if rec.explain:
        result["explain"] = rec.explain

    return result


data = classify_single_url("https://www.itu.int/en/Pages/default.aspx", explain=True)
print(data)
```

```python
from __future__ import annotations

import csv
from pathlib import Path

from urlcheck_smith import SiteClassifier, UrlRecord, check_urls


def enrich_csv_with_url_records(
    input_csv: str | Path,
    output_csv: str | Path,
    *,
    url_column: str = "urls",
    explain: bool = False,
    run_http: bool = False,
) -> None:
    classifier = SiteClassifier(explain=explain, normalize_domain=True)

    records: list[UrlRecord] = []
    input_rows: list[dict[str, str]] = []

    with Path(input_csv).open("r", newline="", encoding="utf-8") as f:
        reader = csv.DictReader(f)
        for row in reader:
            input_rows.append(row)
            url = (row.get(url_column) or "").strip()

            if not url:
                records.append(UrlRecord(url=""))
                continue

            records.append(classifier.classify([UrlRecord(url=url)])[0])

    if run_http:
        records = check_urls(records)

    fieldnames = list(input_rows[0].keys()) if input_rows else []
    fieldnames += [
        "base_url",
        "category",
        "http_status",
        "redirected_url",
        "human_check_suspected",
        "soft_404_detected",
        "trust_tier",
        "error",
        "explain",
    ]

    with Path(output_csv).open("w", newline="", encoding="utf-8") as f:
        writer = csv.DictWriter(f, fieldnames=fieldnames)
        writer.writeheader()

        for row, rec in zip(input_rows, records, strict=False):
            row["base_url"] = rec.base_url or ""
            row["category"] = rec.category or ""
            row["http_status"] = "" if rec.http_status is None else rec.http_status
            row["redirected_url"] = rec.redirected_url or ""
            row["human_check_suspected"] = rec.human_check_suspected
            row["soft_404_detected"] = rec.soft_404_detected
            row["trust_tier"] = rec.trust_tier or ""
            row["error"] = rec.error or ""
            row["explain"] = rec.explain or ""
            writer.writerow(row)


enrich_csv_with_url_records("input.csv", "output.csv", run_http=False)
```
